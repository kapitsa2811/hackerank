```python
features=['countryCode', 'InvoiceAmount','Disputed', 'PaperlessBill', 'DaysToSettle']
```

```python
X = df[features]
y = df['isLate']
y.head(5)
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.30,shuffle=True)
```

```python
print("\n\t X_train:",X_train.shape)
print("\n\t X_test:",X_test.shape)
print("\n\t y_train:",y_train.shape)
print("\n\t y_test:",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
X_test.shape
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python


from sklearn.metrics import accuracy_score

y_pred = logmodel.predict(X_test)

score =accuracy_score(y_test,predictions)
print("\n\t score:",score)
linear_accuracy = round(logmodel.score(X_train,y_train)*100,2)
print("Accuracy:",round(linear_accuracy,2),'%')
```
